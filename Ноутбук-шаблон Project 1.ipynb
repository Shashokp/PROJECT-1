3). Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:

* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def get_experience(arg):
    """
    Функция выделяет из строки число, означающее опыт работы и возвращает число, равное количеству месяцев
    аргументы: 1 строка
    возвращаемое значение: int - количество месяцев, либо NaN
    """
    if type(arg) == float: 
        return np.NaN
    if arg == "Не указано":
        return np.NaN
    arg_list = arg.split(' ')[2:6]
    y, m = 0, 0
    c = []
    for i, val in enumerate(arg_list):
    
        if val in ['лет', 'года', 'год']:
            y = int(arg_list[i-1])
        if val in ['месяц', 'месяца', 'месяцев']: 
            m = int(arg_list[i-1])

    return (y*12+m)

hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(get_experience)
hh_data.drop('Опыт работы', axis=1, inplace=True)
# в качестве проверки выведим список столбцов
hh_data.columns
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
def get_town(arg):
    """
    Функция выделяет из строки название города
    аргументы: 1 строка
    возвращаемое значение: str, название города либо категории городов
    """ 
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    town = arg.split(' ')[0]
    if town in ['Москва', 'Санкт-Петербург']:
        return town
    elif town in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

hh_data['Город'] = hh_data['Город, переезд, командировки'].apply(get_town)

def movability(arg):
    """
    Функция определяет готовность/не готовность к переезду
    аргумент: строка
    возвращаемое значение: True или False
    """
    
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg):
        return False
    elif 'хочу' in arg:
        return True
    else:
        return True
    
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(movability)

def travel_readiness(arg):
    """
    Функция определяет готовность/не готовность к командировкам
    аргумент: строка
    возвращаемое значение: True или False
    """
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or('не готова к командировкам' in arg):
            return False
        else:
            return True
    else:
        return False
    
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(travel_readiness)

hh_data.drop('Город, переезд, командировки', axis=1, inplace=True)
```

```python
#ваш код здесь
currency_rate = pd.read_csv('./data/ExchangeRates.csv', sep=',')

currency_rate['date'] = pd.to_datetime(currency_rate['date'], dayfirst=True)

hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True).dt.date
hh_data['Обновление резюме'] #Не понял почему тип данных - object? Ведь предыдущий код должен был перевести этот признак в тип datetime
# Этот вопрос адресую менторам, которые будут проверять мой код.
# В связи с этим непонятным моментом еще раз изменим тип данных:
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'])

hh_data['желаемая ЗП'] = hh_data['ЗП'].apply(lambda x: x.split(' ')[0])
# Тип данных будет object, как и у родительского признака.
# Поэтому приведем к целому типу
hh_data.loc[:,'желаемая ЗП'] = hh_data['желаемая ЗП'].astype('int')

def get_salary_сurrency(arg):
    """
    функция преобразует наименование валюты к международному сокращенному виду
    аргумент: строка
    возвращаемое значение: str из словаря  
    """
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        'грн': 'UAH', 'белруб': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        'сум': 'UZS', 'AZN': 'AZN'
    }
    curr = arg.split(' ')[1].replace('.', '')
    if curr == 'руб':
        return 'RUB'
    else:
        return currency_dict[curr]  
    
hh_data['валюта ЗП'] = hh_data['ЗП'].apply(get_salary_сurrency)

hh_data_curr = hh_data.merge(
    currency_rate,
    left_on=['валюта ЗП', 'Обновление резюме'],
    right_on=['currency', 'date'],
    how='left'
)

hh_data_curr.loc[hh_data_curr['валюта ЗП']=='RUB', 'close'] = 1 # Значение close для рубля заполнить единицей 1 (курс рубля самого к себе).

# предварительно, перед перемножением признаков, вместо пропусков на Рублях проставим 1 в признаке Пропорция
hh_data_curr.loc[hh_data_curr['валюта ЗП']=='RUB', 'proportion'] = 1

# нужно привести тип данных к целому. До этого тип данных был - object
hh_data_curr.loc[:,'желаемая ЗП'] = hh_data_curr['желаемая ЗП'].astype('int')

hh_data_curr['ЗП (руб)'] = hh_data_curr['желаемая ЗП']*hh_data_curr['close']/hh_data_curr['proportion']

# Чему равна желаемая медианная заработная плата соискателей в нашей таблице (в рублях)?
hh_data_curr['ЗП (руб)'].median()
```
